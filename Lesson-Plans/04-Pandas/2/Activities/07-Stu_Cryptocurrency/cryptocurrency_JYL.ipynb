```python
import pandas as pd
```

```python
# Create a reference the CSV file desired
bitcoin_path = "../07-Stu_Cryptocurrency/Resources/bitcoin_cash_price.csv"

# Read the CSV into a Pandas DataFrame
bitcoin_df = pd.read_csv(bitcoin_path)

# Print the first five rows of data to the screen
bitcoin_df.head()
```

```python
dash_path = "../07-Stu_Cryptocurrency/Resources/dash_price.csv"
dash_df = pd.read_csv(dash_path)
dash_df.head()
```

```python
# Perform an inner merge that combines both DataFrames on the "Date" column
merge_table = pd.merge(bitcoin_df, dash_df, on = "Date")
merge_table.head()
```

```python
# Rename the columns within the newly merged DataFrame so that the headers are more descriptive
merge_table_rename = merge_table.rename(columns = {'Open_x': 'Open bitcoin', 'High_x': 'High bitcoin'
                                            , 'Low_x': 'Low bitcoin', 'Close_x': 'Close bitcoin'
                                           , 'Volume_x': 'Volume bitcoin', 'Market Cap_x': 'Market Cap bitcoin'
                                           , 'Open_y': 'Open dash', 'Close_y': 'Close dash'
                                           , 'Volume_y': 'Volume dash', 'Market Cap_y': 'Market Cap dash'})
merge_table_rename.head()
```
